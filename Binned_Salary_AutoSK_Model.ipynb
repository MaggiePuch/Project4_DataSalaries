Import Dependencies


---



```python
# Import dependencies
import autosklearn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import tensorflow as tf
from sklearn.metrics import classification_report
import autosklearn.classification
import sklearn.model_selection
import sklearn.datasets
import sklearn.metrics
```

Import Data file and Save as Dataframe

---




```python
# Import data file from Google drive
from google.colab import drive
drive.mount('/content/drive')
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/processed_salaries.csv")
df.head()
```

Data Pre-processing


---

Dropping unnecessary columns, converting strings to booleans using Pandas Get Dummies, splitting and scaling the data.

```python
# Drop unused columns
df = df.drop(columns=['salary_in_usd', 'work_year'])
df.head()
```

```python
# Convert categorical data to numeric with 'pd.get_dummies'
df = pd.get_dummies(df, columns=['experience_level',  'employment_type',  'job_title',  'employee_residence', 'company_size'])
df.head()
```

```python
# Split preprocessed data into features and target arrays. Here we make salary_binned as our target array as that is what we would like predicted
y = df['salary_binned'].values

X = df.drop(columns=['salary_binned']).values

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Instantiate StandardScaler instance
scaler = StandardScaler()

# Fit training data to the standard scaler
X_scaler = scaler.fit(X_train)

# Transform training data using the scaler
X_train_scaled = X_scaler.transform(X_train)

# Transform testing data using the scaler
X_test_scaled = X_scaler.transform(X_test)
```

Training models and retreiving accuracy scores


---



```python
# Train the model using AutoSklearn
cls = autosklearn.classification.AutoSklearnClassifier()
cls.fit(X_train_scaled, y_train)
```

```python
# Create predictions using the testing data
y_pred = cls.predict(X_test_scaled)
```

```python
 # Print accuracy score
 print('Test Acc: %.3f' % cls.score(X_test_scaled, y_test))
```

```python
# Print classification report comparing the testing data to the model predictions
print(classification_report(y_test, y_pred))
```

```python
# Show classifier ranking to determine the most effecient classifier used
cls.show_models()
```
